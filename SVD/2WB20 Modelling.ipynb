```python
def DNA_list(n:int)->list:
    """Takes in a value and returns a strand of DNA of length n"""
    DNA = ['A','G','T','C']
    DNA_list=[]
    for i in range(n):
        r = random.randint(0,3)
        DNA_list.append(DNA[r])

    return ''.join(DNA_list)
```

```python
def DNA_list2(n:int)->list:
    """Takes in a value and returns a strand of DNA of length n"""
    DNA = ['A','G','T','C']
    DNA_list=''
    for i in range(n):
        r = random.randint(0,3)
        DNA_list += DNA[r]

    return DNA_list
```

```python
def comb_n(DNA,n)->list:
    """Takes a list of DNA strands ACTG and finds all permutations of n permutations"""
    permutations_list = []

    #Generate all permutations of length 3 
    for perm in product(DNA, repeat=n):
        permutations_list.append(perm)

    return(permutations_list)

```
